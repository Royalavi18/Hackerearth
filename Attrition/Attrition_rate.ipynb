```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Dataset/Train.csv')
```

```python
dataset.head()
```

```python
X=dataset.drop(['Attrition_rate'],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
Label=LabelEncoder()

dataset['Gender']=Label.fit_transform(dataset['Gender'])
```

```python
X['Gender']=dataset['Gender']
```

```python
X
```

```python
X['Relationship_Status']=Label.fit_transform(X['Relationship_Status'])
X['Hometown']=Label.fit_transform(X['Hometown'])
X['Unit']=Label.fit_transform(X['Unit'])
X['Decision_skill_possess']=Label.fit_transform(X['Decision_skill_possess'])
X['Compensation_and_Benefits']=Label.fit_transform(X['Compensation_and_Benefits'])
```

```python
y=dataset['Attrition_rate']
```

```python
X=X.fillna(X.mean())
```

```python
X.isnull().any()
```

```python
X
```

```python
X=X.drop(['Employee_ID'],axis='columns')
```

```python
X.head()
```

```python
y
```

```python
len(X.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
```

```python
scaled_data=scaler.transform(X)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
print(x_pca.shape)
scaled_data.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()

linear_model.fit(x_pca,y)
```

```python
linear_model.intercept_
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_model=DecisionTreeRegressor()

tree_model.fit(x_pca,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model=KNeighborsRegressor()

knn_model.fit(x_pca,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()

rf_model.fit(x_pca,y)
```

```python
from sklearn.svm import SVR
sv_model=SVR(kernel='rbf')

sv_model.fit(x_pca,y)
```

```python
import xgboost as xgb

xg_model = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)

xg_model.fit(x_pca,y)
```
